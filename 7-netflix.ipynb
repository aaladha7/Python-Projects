__Q 10. What is the maximum duration of a Movie/Show on Netflix__

```python
data.head(2)
```

```python
data['Duration'].unique()
```

```python
#we will ignore seasons in this case and only focus on minutes
data['Duration'].dtypes
    #its an Object
```

```python
data.head(2)
```

```python
#str.split() we will split the columns between the ' '
    #first we define the 2 columns we want to create, then we split them with 'space'
        #we split in 'space' because there is a gap between # in duration ie 93 min, 4 season
        #we add expand = True
            #expand=True: Expands the split strings into separate columns in a DataFrame.
            #expand=False (default): Returns the split strings as lists in a Series.
    
data[['Minutes', 'Units']] = data['Duration'].str.split(' ', expand=True)
```

```python
data.head(2)
```

```python
data.dtypes
```

```python
data['Minutes'] = data['Minutes'].astype(int)
#now we will cast Minutes as an Integer
```

```python
data.dtypes
```

```python
data['Minutes'].max()
```

__Q 13.__

__a) Find all the instances where Category is Movie and Type is Dramas__

__b) Find all instances where Category is TV show and Type is 'Kids' TV'__

__c) Combine both condition a OR condition b together__

```python
data.head(2)
```

```python
#a) Find all the instances where Category is Movie and Type is Dramas

data[ (data['Category']=='Movie') & (data['Type']=='Dramas')]
```

```python
#b) Find all instances where Category is TV show and Type is 'Kids' TV'

data.head(2)
```

```python
data[(data['Category']=='TV Show') & (data['Type']=="Kids' TV")]
#Kids' TV has an apostrophe after the s, so we use double quotes when == to it.
```

```python
#c) Combine both condition a OR condition b together
    #a) Find all the instances where Category is Movie and Type is Dramas
    #b) Find all instances where Category is TV show and Type is 'Kids' TV'

data[ (data['Category']=='Movie') & (data['Type']=='Dramas') | (data['Category']=='TV Show') & (data['Type']=="Kids' TV") ]
```
