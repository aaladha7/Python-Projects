__4. What is the mean stop_duration__
 - use mapping 

```python
df.head()
```

```python
df['stop_time'].unique()
#the datatype of stop_duration is an object/strig
#we cannot get the mean of a string
```

```python
df['stop_duration'].value_counts()
```

```python
df['stop_duration'] = df['stop_duration'].map({'0-15 Min' : 7.5, '16:30 Min': 24, '30+ Min': 45})
#in this case we will map those strings to their mid point values.
```

```python
df['stop_duration'].mean()
```
