```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
```

```python

# Start coding here! Use as many cells as you like

netflix_df.head()

df = netflix_df

#subsetting the df to isolate movies between 1990 & 1999
subset = df['release_year'].between(1990,1999)
df = df[subset]

df['release_year'].value_counts()

#creating a histogram to visualize counts of movies in 1990s

    #plt.hist(df['release_year'])
    #plt.xlabel('year')
    #plt.ylabel('frequency of movies')
    #plt.title('distribution of movies released in the 1990s')
    #plt.show()

    #Based on the histogram, netflix had more movies towards the latter half of the 1990s

#using mode to calculate most frequent movie duration in 1990s
df['duration'].value_counts()

df['duration'].mode()


#df['duration'].dtype
    #df['duration'].dtype

### The most common movie duration in the 1990s is 94 minutes.




 


```
